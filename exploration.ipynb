```python
import pandas as pd
from pprint import pprint
import json
import datetime as dt
```

```python
movies_df = pd.read_csv("Resources/tmdb_5000_movies.csv")

movies_df.head()
```

```python
all_genre_names = []

# Iterate through each row in the DataFrame
for index, row in movies_df.iterrows():
    genre_data = json.loads(row['genres'])
    genre_names = []
    for genre in genre_data:
        if "name" in genre:
            genre_names.append(genre["name"])
    genre_names_str = ', '.join(genre_names)
    all_genre_names.append(genre_names_str)
    
all_genre_names
```

```python
movies_df["genres"] = all_genre_names

movies_df
```

```python
all_keywords = []

# Iterate through each row in the DataFrame
for index, row in movies_df.iterrows():
    keyword_data = json.loads(row['keywords'])
    keyword_names = []
    for keyword in keyword_data:
        if "name" in keyword:
            keyword_names.append(keyword["name"])
    keyword_names_str = ', '.join(keyword_names)
    all_keywords.append(keyword_names_str)
    
all_keywords
```

```python
movies_df["keywords"] = all_keywords
```

```python
all_production_companies = []

# Iterate through each row in the DataFrame
for index, row in movies_df.iterrows():
    production_companies_data = json.loads(row['production_companies'])
    production_companies_names = []
    for production_companies in production_companies_data:
        if "name" in production_companies:
            production_companies_names.append(production_companies["name"])
    production_companies_names_str = ', '.join(production_companies_names)
    all_production_companies.append(production_companies_names_str)
    
all_production_companies
```

```python
movies_df["production_companies"] = all_production_companies
```

```python
all_production_countries = []

# Iterate through each row in the DataFrame
for index, row in movies_df.iterrows():
    production_countries_data = json.loads(row['production_countries'])
    production_countries_names = []
    for production_countries in production_countries_data:
        if "name" in production_countries:
            production_countries_names.append(production_countries["name"])
    production_countries_names_str = ', '.join(production_countries_names)
    all_production_countries.append(production_countries_names_str)
    
all_production_countries
```

```python
movies_df["production_countries"] = all_production_countries
```

```python
all_spoken_languages = []

# Iterate through each row in the DataFrame
for index, row in movies_df.iterrows():
    spoken_languages_data = json.loads(row['spoken_languages'])
    spoken_languages_names = []
    for spoken_languages in spoken_languages_data:
        if "iso_639_1" in spoken_languages:
            spoken_languages_names.append(spoken_languages["iso_639_1"])
    spoken_languages_names_str = ', '.join(spoken_languages_names)
    all_spoken_languages.append(spoken_languages_names_str)
    
all_spoken_languages
```

```python
movies_df["spoken_languages"] = all_spoken_languages
```

```python
# Drop columns we don't need
movies_df.drop(columns=['homepage', 'status', 'original_title'], inplace=True)
```

```python
# Move id column to front
column_to_move = 'id'
target_position = 0

# Remove the column and store it in a temporary variable
temp_col = movies_df.pop(column_to_move)

# Insert the column at the target position
movies_df.insert(target_position, column_to_move, temp_col)

# Display the DataFrame to verify the column move
movies_df
```

```python
# Move title column to second spot
column_to_move = 'title'
target_position = 1

# Remove the column and store it in a temporary variable
temp_col = movies_df.pop(column_to_move)

# Insert the column at the target position
movies_df.insert(target_position, column_to_move, temp_col)

# Display the DataFrame to verify the column move
movies_df
```

```python
# Move budget column closer to revenue column
column_to_move = 'budget'
target_position = 10

# Remove the column and store it in a temporary variable
temp_col = movies_df.pop(column_to_move)

# Insert the column at the target position
movies_df.insert(target_position, column_to_move, temp_col)

# Display the DataFrame to verify the column move
movies_df
```

```python
# Move original_language column closer to spoken_languages column
column_to_move = 'original_language'
target_position = 12

# Remove the column and store it in a temporary variable
temp_col = movies_df.pop(column_to_move)

# Insert the column at the target position
movies_df.insert(target_position, column_to_move, temp_col)

# Display the DataFrame to verify the column move
movies_df
```

```python
# Move popularity column closer to vote columns
column_to_move = 'popularity'
target_position = 14

# Remove the column and store it in a temporary variable
temp_col = movies_df.pop(column_to_move)

# Insert the column at the target position
movies_df.insert(target_position, column_to_move, temp_col)

# Display the DataFrame to verify the column move
movies_df
```

```python
# Fill NA values with blanks
movies_df = movies_df.fillna(' ')
```

```python
# Create a weighted_rating column that uses vote_count and vote_average
v = movies_df['vote_count']
R = movies_df['vote_average']

# Calculate the quantile for 'vote_count'
m = v.quantile(0.70)

# Calculate the mean of 'vote_average'
c = R.mean()

# Calculate the weighted rating 'weighted_rating' for each movie
movies_df['weighted_rating'] = (R * v + m * c)/(v + m)
movies_df
```

```python
# Create a profit column based on revenue and budget
movies_df['profit'] = movies_df['revenue'] - movies_df['budget']

# Move profit column after budget and revenue
column_to_move = 'profit'
target_position = 10

# Remove the column and store it in a temporary variable
temp_col = movies_df.pop(column_to_move)

# Insert the column at the target position
movies_df.insert(target_position, column_to_move, temp_col)

# Display the DataFrame to verify the column move
movies_df
```

```python
# List of column names
column_names = movies_df.columns.tolist()
print(column_names)
```
